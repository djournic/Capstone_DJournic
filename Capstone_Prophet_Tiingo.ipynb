I've noticed there's a lot to write to get a simple dataframe of stock prices from tiingo, so I wrote a little function to streamline the process.

```python
def dstock(asset, timeframe='2y'):
    """This function will work with tiingo to quickly get a dataframe of a stock's 
    closing price, with the timeframe specified
    Arguments:
    asset-stock symbol string, all caps(ex: 'CAT','AAPL')
    timeframe- a string, how far back to go, default is 2 years (ex: '1d','2m','3y')
    """
    deltas=['d','m','y']
    today = date.today()
    spl=len(timeframe)-1
    end=today
    if timeframe[-1] not in deltas:
        print('Error:Incorrect format')
        return
    if timeframe[-1]=='d':
        start=int(timeframe[:spl])
        start_date=today-timedelta(days=start)
    elif timeframe[-1]=='m':
        start=int(timeframe[:spl]) *30
        start_date=today-timedelta(days=start)
    elif timeframe[-1]=='y':
        start=int(timeframe[:spl])*365
        start_date=today-timedelta(days=start)
    df=client.get_dataframe([asset],
                            frequency='daily',
                            metric_name='close',
                            startDate=start_date,
                            endDate=today)
    df.columns=['Close']
    df.index = pd.to_datetime(df.index, format = '%Y-%m-%d').strftime('%Y-%m-%d')
    return df
```

```python
def prof(df):
    """This function will take a dataframe of a given stock's closing price data,
    format it and run it through Prophet forecasting."""
    prof_1 = Prophet()
    #df.index.names = ['ds']
    df.columns=['ds','y']
    df.reset_index(drop=True, inplace=True)
    prof_1.fit(df)
    future = prof_1.make_future_dataframe(periods=21)
    forecast = prof_1.predict(future)
    df1_cv = cross_validation(prof_1, initial='30 days', period='7 days', horizon = '21 days')
    df1_pm = performance_metrics(df1_cv)
    
    return forecast,df1_cv,df1_pm
```
